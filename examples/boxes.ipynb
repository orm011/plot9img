```python
import PIL.Image
import plot9img
from plot9img import ggimg, ggimg_draw
```

```python
import os
```

```python
test_path = f'{os.path.dirname(plot9img.__file__)}/../tests/vine_med.jpeg'
```

```python
test_image = PIL.Image.open(test_path)
```

```python
test_image
```

```python
import pandas as pd
# get your annotation data into a dataframe format
annotations_df = pd.DataFrame(
        [{"x1": 100, "x2": 225, "y1": 100, "y2": 300, "label": "A"},
         {"x1": 200, "x2": 325, "y1": 400, "y2": 500, "label": "B"},
         {"x1": 300, "x2": 400, "y1": 200, "y2": 300, "label": "A"}
        ]
)
```

```python
annotations_df
```

```python
from plotnine import geom_rect, geom_text,aes

# standard ggplot geom API
```

```python
# let plotnine map categories to colors, keep consistency across colors, etc. 
tmp = (ggimg(image=test_image, data=annotations_df)
        + geom_rect(
            aes(xmin="x1", xmax="x2", ymin="y1", ymax="y2", color="label"), fill=None, size=2,
            )
        + geom_text(aes(label="label", x="x1+5", y="y1-20", color="label"), 
                        size=13, fontweight='bold', ha='left', va='top')
      )

ggimg_draw(tmp)
```
